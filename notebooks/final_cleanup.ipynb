```python
anime_df = pd.read_csv('../DATA/Anime_data_content.csv')
anime_new = anime_df.copy()
anime_new.head(3)
```

```python
anime_new = anime_new.sort_values(by=['Title'])
```

```python
anime_new= anime_new.reset_index()
anime_new = anime_new.drop(columns='index')
```

```python
anime_new.head()
```

```python
anime_titles = anime_new['Title'].tolist()
```

```python
len(anime_titles)
```

```python
from collections import OrderedDict
def compare(s,t):
    len_s = len(s)
    if s == t[:len_s]:
        return True
    else:
        return False
```

```python
def grouping(titles):
    d = OrderedDict()
    d[titles[0]] = []

    for i in range(1,len(titles)):
        target_key = list(d.keys())[-1]
        if compare(target_key,titles[i]):
            d[target_key].append(titles[i])
        else:
            d[titles[i]] = []
    return list(d.keys())
```

```python
res_titles = grouping(anime_titles)
```

```python
len(res_titles)
```

```python
res_titles
```

```python
anime_new = anime_new[anime_new['Title'].isin(res_titles)]
```
